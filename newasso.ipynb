## กิจกรรม

```python
ass_s1 = df_a['กิจกรรม_2564']
ass_s1
```

```python
import re
def clean_row(row):
    # Remove parentheses and their contents
    row = re.sub(r'\([^)]*\)', '', row)
    # Remove numbers
    row = re.sub(r'[0-9]', '', row)
    return row.strip()
```

```python
# Apply the clean_row function to each row in the list
cleaned_rows = [clean_row(row) for row in ass_s1]
```

```python
cleaned_rows_1 = pd.DataFrame({'กิจกรรม': cleaned_rows})
```

```python
cleaned_rows_2 = cleaned_rows_1['กิจกรรม'].str.split(',', expand=True)
```

```python
df_a['กิจกรรม_2564'] = cleaned_rows_2[0]
```
