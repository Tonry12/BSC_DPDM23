```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# ตั้งค่า DataFrame ใหม่ที่มีเฉพาะคอลัมน์ที่สนใจ
df_association = df_a[['จังหวัด', 'กิจกรรม_2564']]

# แปลงข้อมูลใน DataFrame ให้อยู่ในรูปแบบที่เหมาะสมสำหรับ Apriori algorithm
df_encoded = pd.get_dummies(df_association)

# ใช้ Apriori algorithm เพื่อค้นหากฎการเชื่อมโยง
frequent_itemsets = apriori(df_encoded, min_support=0.025, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)

# แสดงกฎการเชื่อมโยงที่ได้
print(rules)
```
